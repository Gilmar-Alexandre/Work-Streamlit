```python
import pandas as pd
```

```python
df = pd.read_excel('dados.xlsx')
df.head(3)
```

```python
df.info()
```

```python
columns_to_convert = ['INÍCIO', 'TÉRMINO']

for col in columns_to_convert:
    df[col] = pd.to_datetime(df[col], format='%d/%m/%Y', errors='coerce')
```

```python
df.info()
```

```python
df['VALOR PAGO'] = df['VALOR PAGO'].replace('R$', '')
```

```python
# Substituir vírgulas por pontos
df['VALOR PAGO'] = pd.to_numeric(df['VALOR PAGO'], errors='coerce')
df['DIFERENÇA DE VALOR DE CONTRATO'] = pd.to_numeric(df['DIFERENÇA DE VALOR DE CONTRATO'], errors='coerce')
```

```python
fill_values = {
    'CNPJ': '11.111.111/1111-11',
    'SISTEMA': 'nulo',
    'VALOR REAJUSTADO': 0,
    'PEDIDO/ORDEM DE COMPRAS': 'nulo',
    'DATA DA AÇÃO': '01/01/2024',
    'MOTIVO DO CANCELAMENTO': 'nulo',
    'TIPO DO CANCELAMENTO': 'nulo',
    'RESPONSÁVEL PELO CONTRATO': 'nulo',
    'DIFERENÇA DE VALOR DE CONTRATO': 0
}

df = df.fillna(value=fill_values)
```
